```python
import pandas as pd
import json
```

```python
src_file = 'gast_kronplatz_radius_4.json'
```

```python
with open(src_file) as f:
    data = json.load(f)
```

```python
data['Items'][0]['Detail']['en']['Title']
```

```python
data_out = []
```

```python
for item in data['Items']:
    item_out = {'ODH-BZ-Id': item['Id'], 
                'Name': item['Detail']['en']['Title'],
               'Description': item['Detail']['en']['BaseText'],
               'Altitude': item['Altitude'],
               'Latitude': item['Latitude'],
               'Longitude': item['Longitude'],
               'DishRates': item['DishRates']}
    data_out.append(item_out)
```

```python
data_out
```

```python
file_out = 'gast.json'
json_data = {'List': data_out}
```

```python
with open(file_out, 'w') as fp:
    fp.write(json.dumps(json_data))
```
